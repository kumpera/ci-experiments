```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Overview
To produce res.csv:
1. Build my vw branch: https://github.com/kumpera/vowpal_wabbit/tree/safe_reduction
2. Generate the input file `python gen_challenger.py > input.txt` (the script in in the vw repo)
3. call vw: `./vw --cb_explore_adf --cb_challenger  -d input.txt  -q SA --quiet > res.csc`

```python
names = ['idx', 'policy_lowerbounds', 'policy_expectation', 'baseline_lowerbounds', 'baseline_expectation', 'mean_reward', 'mean_w_policy', 'mean_w_baseline']

df = pd.read_csv('res.csv', names=names)
df.describe()
```

```python
plt.figure(figsize=(16,6))
plt.plot(df.policy_lowerbounds, label= 'policy - lower bounds')
plt.plot(df.baseline_lowerbounds, label= 'baseline - lower bounds')
plt.plot(df.policy_expectation, label= 'policy - expectation')
plt.plot(df.baseline_expectation, label= 'baseline - expectation')
plt.legend()
```
